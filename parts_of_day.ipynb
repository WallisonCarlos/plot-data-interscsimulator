```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import EllipseCollection
```

```python
scenario = 0
files = os.listdir('/home/wally/eclipse-workspace/pkmeansimulator/timer/global/'+str(scenario)+'/PARTS_OF_DAY/')
dataFrames = list()
dataFramesNames = list()
for j in range(len(files)):
    print(files[j])
    df = pd.read_csv("/home/wally/eclipse-workspace/pkmeansimulator/timer/global/"+str(scenario)+"/PARTS_OF_DAY/"+str(files[j]), delimiter = ";") 
    dataFrames.append(df)
    dataFramesNames.append(files[j])
```

```python
def chartFirsts(size, df):
    df = df['link'].value_counts()
    df.head()
    df.head(size).plot(kind = 'bar')
    plt.show()
```

```python
def chartLasts(size, df):
    df = df['link'].value_counts()
    df.tail(size).plot(kind = 'bar')
    plt.show()
```
